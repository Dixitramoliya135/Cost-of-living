```python
import pandas as pd
df = pd.read_csv('/content/sample_data/costofliving.csv')
df
```

```python
#findinf ull values
df.isnull().sum()
```

```python
#info of df
df.info()
```

```python
#statical info of df
df.describe()
```

```python
df['total cost leaving'] = (df['Cost of Living Plus Rent Index'] + df['Groceries Index'] + df['Restaurant Price Index'] + df['Local Purchasing Power Index'])
```

```python
df['variance with cost and groceries'] = df['Cost of Living Plus Rent Index'] - df['Groceries Index'] + df['Restaurant Price Index'] + df['Local Purchasing Power Index']
df
```

```python
#split columns
df[['city','country','none']] = df['City'].str.split(',' , expand = True)
```

```python
#delete none column from splitting
del df['none']
```

```python
df
```

```python
#sorting according cost
df.sort_values(by = ['total cost leaving','Cost of Living Plus Rent Index'], ascending = [False,False], inplace = True)
```

```python
#top ten city according to total cost
df[['city','country', 'total cost leaving']].head(10)
```

```python
#filter
high_cost = df['total cost leaving']>200
df.loc[high_cost][['city','country','total cost leaving']]
```

```python
#percetage wise groceries and without groceries cost
df['percentage'] = (df['variance with cost and groceries']/df['total cost leaving'])*100
df.sort_values(by = ['percentage'])
```
